```python
# plotting quantiles
def error_plot_rotations3(figsize):
    colors = ['red','blue','green','orange','brown','purple']
    quantile_colors = ['blue','black','red','black','blue']
    x = np.arange(1,7)
    labels = ['10mm\nx,y,xy','10mm\nz,xz,yz',
              '8mm\nx,y,xy','8mm\nz,xz,yz',
              '6.4mm\nx,y,xy','6.4mm\nz,xz,yz']
    plt.figure(figsize = figsize)
    plt.xticks(x, labels, rotation='horizontal',fontsize=15)
    plt.yticks(fontsize=15)
    for rot_ax in ['x', 'y', 'z', 'xy', 'xz', 'yz']:
        if rot_ax == 'x':
            label = 'x - LR/partition '
        elif rot_ax == 'y':
            label = 'y - AP/phase'
        elif rot_ax == 'z':
            label = 'z - HF/readout'
        else: 
            label = rot_ax
        plt.scatter(-3, -3, lw=0,s = 30, c = colors[ax_to_idx[rot_ax]],alpha = 0.3,marker='o' , label=label)
    plt.plot([-4,-5],[-4,-5],lw=1,c='red',label='median')
    plt.plot([-4,-5],[-4,-5],lw=1,c='black',label='25%,75% quantile')
    plt.plot([-4,-5],[-4,-5],lw=1,c='blue',label='5%,95% quantile')
    e1=error_plot(cost_dict_10_filtered,1,'10mm',0.01, ['x','y'], ax_to_idx_tricubic)
    e2=error_plot(cost_dict_10_1_filtered,1,'10mm',0.01, ['xy'], ax_to_idx_tricubic)
    plt.hlines(np.percentile(np.array(e1+e2),[5,25,50,75,95]),1-0.1,1+0.1,lw=1,color=quantile_colors)
    
    e3=error_plot(cost_dict_10_filtered,2,'10mm',0.01, ['z'], ax_to_idx_tricubic)
    e4=error_plot(cost_dict_10_1_filtered,2,'10mm',0.01, ['yz','xz'], ax_to_idx_tricubic)
    plt.hlines(np.percentile(np.array(e3+e4),[5,25,50,75,95]),2-0.1,2+0.1,lw=1,color=quantile_colors)
    
    e5=error_plot(cost_dict_8_filtered,3,'8mm',0.01, ['x','y'], ax_to_idx_tricubic)
    e6=error_plot(cost_dict_8_1_filtered,3,'8mm',0.01, ['xy'], ax_to_idx_tricubic)
    plt.hlines(np.percentile(np.array(e5+e6),[5,25,50,75,95]),3-0.1,3+0.1,lw=1,color=quantile_colors)

    e7=error_plot(cost_dict_8_filtered,4,'8mm',0.01, ['z'], ax_to_idx_tricubic)
    e8=error_plot(cost_dict_8_1_filtered,4,'8mm',0.01, ['yz','xz'], ax_to_idx_tricubic)
    plt.hlines(np.percentile(np.array(e7+e8),[5,25,50,75,95]),4-0.1,4+0.1,lw=1,color=quantile_colors)
    
    e9=error_plot(cost_dict_6_4_filtered,5,'6.4mm',0.01, ['x','y'], ax_to_idx_tricubic)
    e10=error_plot(cost_dict_6_4_1_filtered,5,'6.4mm',0.01, ['xy'], ax_to_idx_tricubic)
    plt.hlines(np.percentile(np.array(e9+e10),[5,25,50,75,95]),5-0.1,5+0.1,lw=1,color=quantile_colors)
    
    e11=error_plot(cost_dict_6_4_filtered,6,'6.4mm',0.01, ['z'], ax_to_idx_tricubic)
    e12=error_plot(cost_dict_6_4_1_filtered,6,'6.4mm',0.01, ['yz','xz'], ax_to_idx_tricubic)
    plt.hlines(np.percentile(np.array(e11+e12),[5,25,50,75,95]),6-0.1,6+0.1,lw=1,color=quantile_colors)
    
    plt.hlines(0,0,7,color = 'grey',lw=1,linestyle='--')
    plt.xlim([0.5,6.5])
    plt.ylim([-0.5,0.4])
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5),fontsize=15)
    plt.xlabel('a',fontsize=15)
    plt.grid('off')
    plt.show()
```
