```python
def normalize(volume, scale):
    assert((volume >= 0).all())
    return volume * scale
```

```python
# def RMS(dRM, dt, rad, res):
#     rad = 100
#     t = dt*res
#     return np.arccos((np.trace(dRM)-1)/2.)

def RMS(dRM, dt, rad, res):
    rad = 100
    A = dRM-np.eye(3)
    t = dt*res
    return np.sqrt(0.2*rad**2*np.trace((A.T).dot(A))+ (t.T).dot(t))
    
from conversion import rotation_matrix_fromq, rotation_matrix_from_params
def get_true_params(rot_angle,rot_ax):
    wy,wx,wz = ax_to_w(rot_ax)
    wy_t,wx_t,wz_t = ax_to_w(axes_dict[rot_ax])
    true_RM = rotation_matrix_fromq(rot_angle[0],wx,wy,wz)
    trans_y, trans_x, trans_z = wy_t*rot_angle[1],wx_t*rot_angle[1], wz_t*rot_angle[1]
    return [trans_x, trans_y, trans_z], true_RM

def get_params(errors, Ps, res):
    params = Ps[-1]
    RM = rotation_matrix_from_params(params[3:])
    return  params[:3]*res, RM
```

```python
import multiprocessing
dict_ = oct13_dict_10
def run_GN_Bsplines(rot_ax):
    print rot_ax,
    cost_dict = {}
    for rang in ranges:
        print rang,
        ref = rep_to_angle(0,rang)
        vol_ref = fourier_filter(dict_[rot_ax][axes_dict[rot_ax]][rang][ref],rad)
        vol_ref_masked = sphere_mask(vol_ref,rad)
        vol_ref_dervs = get_gradient_P(vol_ref, axis_derivatives=BSpline_derivatives)
        for rep in xrange(1,36):
            rot_angle = rep_to_angle(rep,rang)
            print rot_angle,
            vol_m = fourier_filter(dict_[rot_ax][axes_dict[rot_ax]][rang][rot_angle],rad)
            vol_m_dervs = BSpline_coefficients(vol_m)
            cost_dict[rot_angle] = Gauss_Newton(vol_ref_masked, vol_ref_dervs, vol_m, vol_m_dervs, mask=True, 
                                                 divide_factor = 1., alpha = 1., max_iter = 20, 
                                                 plot = False, decrease_factor = 0.25,interp = Bspline_interp)
    return cost_dict
```

```python
def error_plot(cost_dict, col, xlim = [0,5]):
    colors = ['red','blue','green','orange','brown','purple']
    ranges = ['0_5_to_2_5','3_0_to_5_0']
    for i in xrange(6):
        plt.scatter(cost_dict[i*70:(i+1)*70],np.zeros(70)+col,lw=0,s = 30, c = colors[i],alpha = 0.1,marker='o')
        plt.xlim(xlim)
        plt.ylim([0,10])
```

```python
# plotting quantiles
def error_plot_rotations(figsize):
    colors = ['red','blue','green','orange','brown','purple']
    quantile_colors = ['blue','black','red','black','blue']
    y = np.arange(1,10)
    labels = ['10mm\nCubic\nB-spline','10mm\nTricubic','10mm\nTrilinear',
              '8mm\nCubic\nB-spline','8mm\nTricubic','8mm\nTrilinear',
              '6.4mm\nCubic\nB-spline','6.4mm\nTricubic','6.4mm\nTrilinear']
    plt.figure(figsize = figsize)
    plt.yticks(y, labels, rotation='horizontal',fontsize=18)
    plt.xticks(fontsize=18)
    for idx, rot_ax in enumerate(axes_dict.keys()):
        if rot_ax == 'x':
            label = 'x - LR/partition '
        elif rot_ax == 'y':
            label = 'y - AP/phase'
        elif rot_ax == 'z':
            label = 'z - HF/readout'
        else: 
            label = rot_ax
        plt.scatter(-3, -3, lw=0,s = 30, c = colors[idx],alpha = 0.3,marker='o' , label=label)
    plt.plot([-4,-5],[-4,-5],lw=1,c='red',label='median')
    plt.plot([-4,-5],[-4,-5],lw=1,c='black',label='25%, 75% quantile')
    plt.plot([-4,-5],[-4,-5],lw=1,c='blue',label='5%, 95% quantile')

    idx = 1
    error_plot(RMS_ls_bspline_10,idx)
    plt.vlines(np.percentile(np.array(RMS_ls_bspline_10),[5,25,50,75,95]),idx-0.1,idx+0.1,lw=1,color=quantile_colors)
    
    idx += 1 
    error_plot(RMS_ls_cubic_10,idx)
    plt.vlines(np.percentile(np.array(RMS_ls_cubic_10),[5,25,50,75,95]),idx-0.1,idx+0.1,lw=1,color=quantile_colors)

    idx += 1
    error_plot(RMS_ls_linear_10,idx)
    plt.vlines(np.percentile(np.array(RMS_ls_linear_10),[5,25,50,75,95]),idx-0.1,idx+0.1,lw=1,color=quantile_colors)
    
    idx += 1
    error_plot(RMS_ls_bspline_8,idx)
    plt.vlines(np.percentile(np.array(RMS_ls_bspline_8),[5,25,50,75,95]),idx-0.1,idx+0.1,lw=1,color=quantile_colors)
    
    idx += 1
    error_plot(RMS_ls_cubic_8,idx)
    plt.vlines(np.percentile(np.array(RMS_ls_cubic_8),[5,25,50,75,95]),idx-0.1,idx+0.1,lw=1,color=quantile_colors)

    idx += 1
    error_plot(RMS_ls_linear_8,idx)
    plt.vlines(np.percentile(np.array(RMS_ls_linear_8),[5,25,50,75,95]),idx-0.1,idx+0.1,lw=1,color=quantile_colors)

    idx += 1
    error_plot(RMS_ls_bspline_6_4,idx)
    plt.vlines(np.percentile(np.array(RMS_ls_bspline_6_4),[5,25,50,75,95]),idx-0.1,idx+0.1,lw=1,color=quantile_colors)
    
    idx += 1
    error_plot(RMS_ls_cubic_6_4,idx)
    plt.vlines(np.percentile(np.array(RMS_ls_cubic_6_4),[5,25,50,75,95]),idx-0.1,idx+0.1,lw=1,color=quantile_colors)
    
    idx += 1
    error_plot(RMS_ls_linear_6_4,idx)
    plt.vlines(np.percentile(np.array(RMS_ls_linear_6_4),[5,25,50,75,95]),idx-0.1,idx+0.1,lw=1,color=quantile_colors)
    
    plt.vlines(0,0,7,color = 'r',lw=1,linestyle='--')
    plt.xlim([0,0.2])
    plt.ylim([0,10])
    plt.legend(loc='center left', bbox_to_anchor=(0.3, 1.2))
    plt.title('RMS Error Comparison',fontsize = 20)
    plt.xlabel('RMS Error (mm)',fontsize=20)
    plt.grid('off')
    plt.show()
```
